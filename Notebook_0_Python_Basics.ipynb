## Variables — Your First Observations

A **variable** is a name that stores a value. Think of it as labeling a sample in the field.

```python
# Text values are called "strings" — wrap them in quotes
station_name = "Miami International Airport"
station_id = "USW00012839"

# Numbers without decimals are "integers"
elevation_m = 3

# Numbers with decimals are "floats"
latitude = 25.793
longitude = -80.290
mean_temp_c = 25.5

print("Station:", station_name)
print("ID:", station_id)
print("Location:", latitude, "°N,", longitude, "°W")
print("Elevation:", elevation_m, "m")
print("Mean Temperature:", mean_temp_c, "°C")
```

## Lists — Your Station Checklist

A **list** holds multiple values in order, like a checklist of field sites.

```python
# Five US cities we'll study throughout this course
stations = ["Miami", "New York", "Chicago", "Los Angeles", "Seattle"]
avg_temps = [25.5, 13.0, 10.5, 18.0, 11.5]

print("Number of stations:", len(stations))
print("First station:", stations[0])
print("Last station:", stations[-1])

# Add a new station
stations.append("Phoenix")
avg_temps.append(23.5)
print("Updated list:", stations)
```

## Dictionaries — Station Metadata

A **dictionary** maps keys to values, like a field record card.

```python
# One station's metadata
miami = {
    "name": "Miami International Airport",
    "id": "USW00012839",
    "lat": 25.793,
    "lon": -80.290,
    "elevation_m": 3,
    "climate_zone": "Tropical"
}

print(miami["name"])
print("Climate zone:", miami["climate_zone"])
print("Coordinates:", miami["lat"], miami["lon"])
```

```python
# Multiple stations in a list of dictionaries
all_stations = [
    {"name": "Miami",       "temp": 25.5, "zone": "Tropical"},
    {"name": "New York",    "temp": 13.0, "zone": "Continental"},
    {"name": "Chicago",     "temp": 10.5, "zone": "Continental"},
    {"name": "Los Angeles", "temp": 18.0, "zone": "Mediterranean"},
    {"name": "Seattle",     "temp": 11.5, "zone": "Marine"},
]

for s in all_stations:
    print(f"{s['name']:15s}  {s['temp']:5.1f} °C  ({s['zone']})")
```

## Loops — Processing Multiple Stations

A **loop** repeats an action for every item in a list. Instead of writing the same code 5 times, write it once.

```python
stations = ["Miami", "New York", "Chicago", "Los Angeles", "Seattle"]
avg_temps = [25.5, 13.0, 10.5, 18.0, 11.5]

# zip() pairs two lists together element-by-element
for city, temp in zip(stations, avg_temps):
    print(f"{city}: {temp} °C")
```

## Conditionals — Classifying Climates

Use **if/elif/else** to make decisions based on data.

```python
def classify_climate(temp):
    """Classify a station by its annual average temperature."""
    if temp >= 22:
        return "Tropical/Subtropical"
    elif temp >= 15:
        return "Warm Temperate"
    elif temp >= 10:
        return "Cool Temperate"
    else:
        return "Cold"

# Apply to all stations
for city, temp in zip(stations, avg_temps):
    zone = classify_climate(temp)
    print(f"{city}: {temp} °C → {zone}")
```

## Functions — Reusable Tools

A **function** is a reusable block of code. Define it once, use it anywhere.

```python
def celsius_to_fahrenheit(c):
    """Convert Celsius to Fahrenheit."""
    return c * 9/5 + 32

def compute_anomaly(temp, baseline):
    """Compute how far a temperature is from a baseline (long-term average)."""
    return temp - baseline

# Test our functions
baseline = sum(avg_temps) / len(avg_temps)  # mean of all stations
print(f"Baseline (mean of all stations): {baseline:.1f} °C")
print()

for city, temp in zip(stations, avg_temps):
    f = celsius_to_fahrenheit(temp)
    anom = compute_anomaly(temp, baseline)
    sign = "+" if anom >= 0 else ""
    print(f"{city}: {temp} °C = {f:.1f} °F  (anomaly: {sign}{anom:.1f} °C)")
```

## Your First Plot

Let's visualize our station data with a **bar chart**.

```python
import matplotlib.pyplot as plt

stations = ["Miami", "New York", "Chicago", "Los Angeles", "Seattle"]
avg_temps = [25.5, 13.0, 10.5, 18.0, 11.5]
colors = ["#e74c3c", "#3498db", "#2ecc71", "#f39c12", "#9b59b6"]

plt.figure(figsize=(10, 5))
plt.bar(stations, avg_temps, color=colors)
plt.title("Annual Average Temperature by City")
plt.xlabel("City")
plt.ylabel("Temperature (°C)")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

## Sine Wave vs. Real Climate

In geoscience, we often compare **idealized models** to **real observations**. Let's see how a simple sine wave compares to Miami's actual monthly temperatures.

```python
import numpy as np

months = np.arange(1, 13)
month_names = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Idealized seasonal cycle: sine wave centered at 25°C, amplitude 5°C
ideal = 25 + 5 * np.sin(2 * np.pi * (months - 4) / 12)

# Real Miami monthly averages (approximate)
real_miami = [20.1, 20.8, 22.3, 24.1, 26.3, 27.8, 28.5, 28.6, 27.8, 25.9, 23.1, 21.0]

plt.figure(figsize=(10, 5))
plt.plot(months, ideal, "b--o", label="Idealized Sine Wave", alpha=0.7)
plt.plot(months, real_miami, "r-s", label="Real Miami Data", linewidth=2)
plt.xticks(months, month_names)
plt.title("Idealized vs. Real Seasonal Cycle (Miami)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Notice: real climate is messier than the idealized model!")
print("The real data has asymmetry — warming is slower than cooling.")
```
